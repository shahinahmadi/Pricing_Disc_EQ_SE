# Data matching between the saved price list and new price list for UPH

```python
import pandas as pd
import numpy
import csv
from itertools import zip_longest

data1=pd.read_excel("Pricing_Discrepancy.xlsx", sheet_name="UPH")
data2=pd.read_excel("Pricing_Discrepancy.xlsx", sheet_name="CPQ Pricing UPH")

skus1=[]
grade11=[]


skus2=[]
grade22=[]


SKUS1=data1['SKU']
SKUS2=data2['INFCUR']

grade1=data1["Cover"]
grade2=data2["INFGRD"]


Price2=data2["INFPRC"]
price2=[]

for i in SKUS1:
    skus1.append(i)

for i in SKUS2:
    skus2.append(i)

for i in grade1:
    grade11.append(i)
    
for i in grade2:
    grade22.append(i)

for i in Price2:
    price2.append(i)

a=[skus1,grade11]
b=[skus2,grade22]
c=price2
extracted_price=[]
extracted_skus=[]
extracted_grades=[]
length1=len(a[0])
length2=len(b[0])


print("----------------------------------------")
print("The length of column in our Excel file: ")
print(length1)
print("----------------------------------------")


print("---------------------------------------------------")
print("The length of the column in the file made for CPQ: ")
print(length2)
print("---------------------------------------------------")



for j in range(0,length1,1):
    for k in range(0,length2,1):
        if ((a[0][j])==b[0][k]) and (a[1][j]==b[1][k]):
            print(b[0][k])
            print(b[1][k])
            print(c[k])
            extracted_skus.append(b[0][k])
            extracted_grades.append(b[1][k])
            extracted_price.append(c[k])
            
            
            
print("-----------------------------------------------")
print("This is the length of the desired price list:  ")
print(len(extracted_skus))
print("-----------------------------------------------")


print("-----------------------------------------------")
print("This is the length of the desired price list:  ")
print(len(extracted_grades))
print("-----------------------------------------------")


print("***********************************************")
print("This is the length of the desired price list:  ")
print(len(extracted_price))
print("***********************************************")



# Exporting the extracted data to the spreadsheet

d=[extracted_skus,extracted_grades,extracted_price]
export_data=zip_longest(*d, fillvalue='')
with open("Extracted.csv", "w", encoding="ISO-8859-1", newline='') as CPQ:
    wr=csv.writer(CPQ)
    wr.writerow(("SKU","Grade","CPQ Price"))
    wr.writerows(export_data)
CPQ.close()   
    
    


    
```
